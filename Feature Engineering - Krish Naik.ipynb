## Missing Values

```python
features_with_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum() >0]

for feature in features_with_nan:
    print(f"The feature is {feature} and missing values in {np.round(dataset[feature].isnull().mean()*100,4)}%")
```

```python
## replacing categorical features

categorical_features = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
```

```python
#1st technique

data = dataset.copy()
def replace_cat_features(dataset, categorical_features):
    data[categorical_features] = data[categorical_features].fillna("Missing")
    return data
```

```python
replace_cat_features(data, categorical_features)
```

```python
#2nd technique

data = dataset.copy()
def replace_cat_features_with_mode(dataset, categorical_features):
    data[categorical_features] = data[categorical_features].fillna(data[categorical_features].mode())
    return data
```

```python
replace_cat_features_with_mode(data,categorical_features)
```

```python
## numerical features with nan

numerical_features_with_nan = [feature for feature in dataset.columns if dataset[feature].dtypes!='object' and dataset[feature].isnull().sum()>=1]
```

```python
numerical_features_with_nan
```

```python
for feature in numerical_features_with_nan:
    ##since there are outlier, we are using median for replacement
    median_values = dataset[feature].median()
    ## create a new feature to capture nan value
    dataset[feature+'nan'] = np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_values, inplace=True)
    
```

The reason we created new feature columns because in Machine Learning when we will apply the algorithms we have the columns that gives the importance of the missing values. It will train our model more efficiently.

```python
## Temporal Variables(datetime variables)

for feature in ['YearBuilt', 'GarageYrBlt', 'YearRemodAdd']:
    dataset[feature +'age of house'] = dataset['YrSold']-dataset[feature]
    
```
