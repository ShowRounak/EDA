```python
## Let's plot some amazing diagram

data = dataset.copy()

for feature in features_with_nan:
    #let's make a variable that indicates 1 if the observation was missing or otherwise
    data[feature]= np.where(data[feature].isnull(),1,0)

    #let's calculate the mean SalePrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot(kind = 'bar')
    plt.title(feature)
    plt.show
```

Here with the relation between the missing values and the dependent variable is clearly visible. So, we need to replace these nan values with something meaningful which we will do in the Feature Engineering section.

### Numerical Variables

```python
dataset.info()
```

```python
# List of numerical variables

dataset['SalePrice'].dtypes != 'object'
```

```python
numerical_features = [feature for feature in dataset.columns if dataset[feature].dtypes!='object']
```

```python
numerical_features
```

```python
len(numerical_features)
```

```python
dataset[numerical_features].head()
```

```python
## Temporal Variables (Eg: Datetime Variables)

year_feature = [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
```

```python
year_feature
```

```python
for feature in year_feature:
    print(feature, dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()

plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('House Price vs Year Sold')
```

```python
## Here we will compare the difference between all years features with SalePrice

data = dataset.copy()
for feature in year_feature:
    if feature != 'YrSold':
        data[feature] = data['YrSold']-data[feature]
        
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
        
```

#### Observation:

Within all the yearwise information, it is clearly visible that newer houses are costlier than older houses. And it looks like a Pareto or Power-law distrbution.

```python
### Numerical variables - 2 types
## 1. Continuous Variable and 2. Discrete Variable

discrete_feature = [feature for feature in numerical_features if len(dataset[feature].unique())<=25] #25 is a threshold set by domain expert

print(len(discrete_feature))
```

```python
dataset[discrete_feature].head()
```

```python
## Lets find out the relationship between discrete and SalePrice

data = dataset.copy()
for feature in discrete_feature:
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
    
```

#### Observation:

there is a relationship between Discrete features and SalePrice

```python
## Continous Variable

continous_feature = [feature for feature in numerical_features if feature not in discrete_feature+year_feature] 

print(len(continous_feature))
```

```python
### let's analyse the continous values by creating histogram to understand the distribution

data  = dataset.copy()

for feature in continous_feature:
    data.groupby(feature)['SalePrice'].median().plot.hist()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
data  = dataset.copy()

for feature in continous_feature:
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

## EDA Part 2

```python
## We will be using logarithmic transformation

data = dataset.copy()

for feature in continous_feature:
    if 0 in data[feature].unique(): #beacuse log(0) is undefined
        pass
    else:
        data[feature] = np.log(data[feature])
        plt.scatter(data[feature], data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.title(feature)
        plt.show()
```

```python
##outliers

for feature in continous_feature:
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column = feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```
