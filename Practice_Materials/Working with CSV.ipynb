## Convertors

```python
pd.read_csv('IPL Matches 2008-2020.csv')
```

Now, I want to change the name of all teams from longname to shortname (Eg: Kolkata Knight Riders -> KKR).

To do this, I need a separate function.

```python
def rename(name):
    if name == "Kolkata Knight Riders":
        return "KKR"
    else:
        return name
```

```python
pd.read_csv('IPL Matches 2008-2020.csv', converters={'winner':rename})
```
