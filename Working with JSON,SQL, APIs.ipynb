## Working with API

```python
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=842ae96f17e1865b0d29c0c53ee62ad1&language=en-US&page=1')
```

```python
response.json()['results']
```

```python
pd.DataFrame(response.json()['results'])
```

Now this is only the data from first page. There are total of 543 pages. So we have to run a loop.

```python
df = pd.DataFrame() #first create an empty dataframe
```

```python
for i in range(1,544):
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=842ae96f17e1865b0d29c0c53ee62ad1&language=en-US&page={i}')
    temp_df = pd.DataFrame(response.json()['results'])
    df = df.append(temp_df, ignore_index=True)
```
